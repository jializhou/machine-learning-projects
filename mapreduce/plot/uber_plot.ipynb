```python
import pandas as pd
import os
import matplotlib  
import matplotlib.pyplot as plt 
%matplotlib inline 
def read_txt(path):
	frames = pd.DataFrame()
	for file in os.listdir(path):
	    if file.endswith("may14.csv"):	    	
	    	file = os.path.join(path,file)
	        frames = frames.append(pd.read_csv(file, sep = ','))
	return frames

df1 = read_txt('uber-2014/')
print df1.dtypes
```

```python
def read_txt(path):
	frames = pd.DataFrame()
	for file in os.listdir(path):
	    if file.endswith("sep14.csv"):	    	
	    	file = os.path.join(path,file)
	        frames = frames.append(pd.read_csv(file, sep = ','))
	return frames

df2 = read_txt('uber-2014/')
print df2.dtypes
```

```python
pd.options.display.mpl_style = 'default' #Better Styling  
new_style = {'grid': False} #Remove grid  
matplotlib.rc('axes', **new_style)  
from matplotlib import rcParams  
rcParams['figure.figsize'] = (17.5, 17) #Size of figure  
rcParams['figure.dpi'] = 250




```

```python
result1 = pd.merge(df1, df2, how = 'outer',on='Date/Time')
result1.head(10)
```

```python
P=result1.plot(kind='scatter', x='Lon_x', y='Lat_x',color='red',xlim=(-74.06,-73.77),ylim=(40.61, 40.91),s=.5,alpha=0.2,title="Uber of NY From April to September in 2014")
result1.plot(kind='scatter', x='Lon_y', y='Lat_y',color='blue',xlim=(-74.06,-73.77),ylim=(40.61, 40.91),s=.5,alpha=0.2,ax=P)
P.set_axis_bgcolor('white') #Background Color
```
