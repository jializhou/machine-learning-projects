```python
import pandas as pd
import os
import matplotlib  
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python

def read_txt(path):
	frames = pd.DataFrame()
	for file in os.listdir(path):
	    if file.endswith("spring_latlong.csv"):	    	
			file = os.path.join(path,file)
			frames = frames.append(pd.read_csv(file, sep = ' '))
			break
	return frames

df1 = read_txt('yellow/')
print df1.dtypes

```

```python
df1.columns = ['Lat', 'Lon']
df1.head()
```

```python
pd.options.display.mpl_style = 'default' #Better Styling  
new_style = {'grid': False} #Remove grid  
matplotlib.rc('axes', **new_style)  
from matplotlib import rcParams  
rcParams['figure.figsize'] = (17.5, 17) #Size of figure  
rcParams['figure.dpi'] = 250



```

```python
P=df1.plot(kind='scatter', x='Lon', y='Lat',color='red',xlim=(-74.06,-73.77),ylim=(40.61, 40.91),s=.5,alpha=0.2,title="Yellow Taxi of NY in Spring")
P.set_axis_bgcolor('white') #Background Color
```
